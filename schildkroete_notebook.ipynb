# Schildkröten-Zeichnungen mit matplotlib

In diesem Notebook können wir Grafiken direkt im Editor sehen!

```python
import matplotlib.pyplot as plt
import numpy as np

# Einfaches Beispiel: Eine Linie zeichnen
fig, ax = plt.subplots(figsize=(6, 6))

# Startpunkt
x = [0, 0]
y = [0, 1]

# Linie zeichnen (wie forward(100) bei Turtle)
ax.plot(x, y, 'b-', linewidth=2)
ax.plot(0, 0, 'ro', markersize=10)  # Startpunkt markieren

# Achsen formatieren
ax.set_xlim(-2, 2)
ax.set_ylim(-2, 2)
ax.set_aspect('equal')
ax.grid(True, alpha=0.3)
ax.set_title('Meine erste Linie')

plt.show()
```

```python
# Ein Quadrat zeichnen
fig, ax = plt.subplots(figsize=(6, 6))

# Quadrat-Koordinaten
x = [0, 1, 1, 0, 0]
y = [0, 0, 1, 1, 0]

ax.plot(x, y, 'g-', linewidth=2)
ax.plot(0, 0, 'ro', markersize=10)  # Startpunkt

ax.set_xlim(-0.5, 1.5)
ax.set_ylim(-0.5, 1.5)
ax.set_aspect('equal')
ax.grid(True, alpha=0.3)
ax.set_title('Ein Quadrat')

plt.show()
```

```python
# Einen Kreis zeichnen
fig, ax = plt.subplots(figsize=(6, 6))

# Kreis mit parametrischen Gleichungen
theta = np.linspace(0, 2*np.pi, 100)
radius = 1
x = radius * np.cos(theta)
y = radius * np.sin(theta)

ax.plot(x, y, 'purple', linewidth=2)
ax.plot(radius, 0, 'ro', markersize=10)  # Startpunkt

ax.set_xlim(-1.5, 1.5)
ax.set_ylim(-1.5, 1.5)
ax.set_aspect('equal')
ax.grid(True, alpha=0.3)
ax.set_title('Ein Kreis')

plt.show()
```
